```python
import pandas as pd
import re
```

Se carga la data

```python
data0=pd.read_csv('dataUN - HDR21-22_Statistical_Annex_HDI_Table.csv')
```

```python
data0
```

Se limita el espacio donde se trabajará

```python
data1=data0.iloc[3:229]
```

```python
data1=data1.reset_index(drop=True) #Se reinicia el index
```

```python
data1
```

Se homogeiniza los nombres de los campos, se eliminan las filas sin ningún valor y las columnas sin importancia 

```python
data2=data1.dropna(axis=0, how='all') #Se elimina las filas sin ningun valor
```

```python
data2=data2.iloc[:,[0,1,2,4,6,8,10,12,14]] #Se considera solo las columnas con importancia
```

```python
data2.columns=data2.iloc[0:3,:].fillna('').apply(lambda x: ' '.join((x))) #Se crea los nombres de las columnas
```

```python
data2.columns=data2.columns.str.strip() #Se quitan espacios inncesarios
```

```python
data3=data2.copy()
```

```python
data3.drop([0,1,2], inplace=True) #Se eliminan filas inncesarias
```

```python
data3
```

```python
data3.reset_index(drop=True, inplace=True) #Se reinicia index
```

Se convierte a números

```python
data4=data3.copy()
```

```python
#Se asegura de que solo hayan símbolos en las columnas númericas de  punto, coma y numeros
data4.loc[:, data4.columns != 'Country']=data4.loc[:, data4.columns != 'Country'].applymap(lambda x: ''.join(re.findall(r'[\d.,]', str(x))))
```

```python
#Se itera por columna y se va coercionando para que tenga valor numerico
for column in data4.columns:
    if column != 'Country':
        data4[column] = pd.to_numeric(data4[column], errors='coerce')
```

```python
data4.info()
```

Se separan a los países individuales de otros paises o territorios

```python
#Encontrar final (191)
paises=data4.iloc[:data4[data4.loc[:,'HDI rank']==191].index[0]+1,:]
```

```python
paises
```

```python
otros=data4.iloc[data4[data4.loc[:,'HDI rank']==191].index[0]+2:,:]
```

```python
otros
```

## Países

```python
#Se trabaja con paises
```

```python
paises['Development']=''
```

```python
categorias=paises[paises['HDI rank'].isna()].Country.reset_index(drop=True)
```

```python
categorias
```

```python
indices=paises[paises['HDI rank'].isna()].Country.index
```

```python
indices
```

```python
paises1=paises.copy()
```

```python
for i in range(len(categorias)): #Se forma columna asignando a que tipo de desarrollo  tiene el pais
    if i==len(categorias)-1:
        paises1.loc[indices[i]: , 'Development']=categorias[i]
    else:
        paises1.loc[indices[i]:indices[i+1], 'Development']=categorias[i]
```

```python
paises2 = paises1[~paises1['Country'].isin(categorias)]
```

```python
paises2
```
